**O que são dataframes?**

São tabelas de duas dimensões que podem ter seus dados observados. Cada coluna tem um tipo somente e um dataframe pode ter vários tipos de colunas.
Um dataframe pode ter:
* outro dataframe
* Uma série
* um array de duas dimensões
* um dicionário de duas dimensões ou listas
* Um ndarray Numpy

**Uso do Concat com Dataframes**
Concatena um dataframe com outro simplesmente emendando.




```python
import pandas as pd # para usar os dataframes

#criando um dicionário
dummy_data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}

#criando um dataframe a partir de um dicionário
df1 = pd.DataFrame(dummy_data1, columns = ['id', 'Feature1', 'Feature2'])

#criando dataframe usando dicionário
dummy_data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
df2 = pd.DataFrame(dummy_data2, columns = ['id', 'Feature1', 'Feature2'])

#criando dataframe usando dicionário
dummy_data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
df3 = pd.DataFrame(dummy_data3, columns = ['id', 'Feature3'])

#usando o concat para juntar dois dataframes
df_row = pd.concat([df1, df2])

print(df1)
print(df2)
print(df_row)
```

A coluna de index está errada. Para ajustá-la deve-se usar o comando

ignore_index=True

```python
df_row_reindex = pd.concat([df1, df2], ignore_index=True)

#concat faz uma cópia completa dos dados e insere em outra variável

df_row_reindex
```

Para rotular que dados vieram de que dataframe, usar a opção keys=['x', 'y']

```python
frames = [df1,df2]
df_keys = pd.concat(frames, keys=['x', 'y'])

df_keys
```

Para definir os rótulos dos dataframes, use a opção keys e passe uma lista de strings.

```python
frames = [df1,df2]
df_keys = pd.concat(frames, keys=['x', 'y'])

df_keys
```

Através dos rótulos das colunas, é possível tb identificar um conjunto de linhas do datraframe.

```python
df_keys.loc['y']
```

Também é possível passar um dicionário para ser usado como rótulo junto com a função concat.

```python
pieces = {'x': df1, 'y': df2}

df_piece = pd.concat(pieces)

df_piece
```

Para concatenar dataframes em uma colunas paralelas e não linhas, insira o parâmetro axis = 1.

```python
df_col = pd.concat([df1,df2], axis=1)
df_col
```

**MERGE**


O comando merge também 'junta' as tabelas, mas usando como referência uma coluna específica para fazer a indexação. Neste caso, o nome da coluna é id.

```python
df_merge_col = pd.merge(df_row, df3, on='id')

df_merge_col
```

Quando as colunas de diferentes dataframes tem nomes diferentes, usa-se os campos left_on e right_on.

```python
df_merge_difkey = pd.merge(df_row, df3, left_on='id', right_on='id')
df_merge_difkey
```

**JOIN**

O comando JOIN é bastante similar ao comando de SQL.

*Full outer join*

Combina os resultados das colunas da direita e da esquerda. A especificação deste tipo de join é feito pela opção how.
Onde não tiver dados, preenche-se com NaNs.

```python
df_outer = pd.merge(df1, df2, on='id', how='outer')

df_outer
```

É possível alterar o sufixo do dataframe usando a opção suffixes.


```python
df_suffix = pd.merge(df1, df2, left_on='id',right_on='id',how='outer',suffixes=('_left','_right'))

df_suffix
```

*INNER JOIN*

Devolve apenas o conjunto interseção entre os dois dataframes.

```python
df_inner = pd.merge(df1, df2, on='id', how='inner')
df_inner
```

```python
#http resquest erros
# A deliberate typo is made in the endpoint "postz" instead of "posts"
url = "https://jsonplaceholder.typicode.com/postz"

# Attempt to GET data from provided endpoint
try:
    response = requests.get(url)
    response.raise_for_status()
# If the request fails (404) then print the error.
except requests.exceptions.HTTPError as error:
  print(error)

"""
404 Client Error: Not Found for url: https://jsonplaceholder.typicode.com/postz
"""
```

```python
#dealing with too many redirects
"""
Note: The code here will not raise an error
but the structure is how you would hand a case where there
are multiple redirects
"""

url = "https://jsonplaceholder.typicode.com/posts"

try:
  response = requests.get(url)
  response.raise_for_status()
except requests.exceptions.TooManyRedirects as error:
  print(error)
```

```python
# Solution 2
url = "https://jsonplaceholder.typicode.com/posts"
session = requests.Session()
session.max_redirects = 3
response = session.get(url)
```

```python
# Solution 3
url = "https://jsonplaceholder.typicode.com/posts"
session = requests.Session()
session.allow_redirects = False
response = session.get(url)
```

```python
#http requests connection errors

"""
Note: The code here will not raise an error
but the structure is how you would hand a case where there
is a connection error.
"""

url = "https://jsonplaceholder.typicode.com/posts"

try:
  response = requests.get(url)
except requests.ConnectionError as error:
  print(error)
```

```python
#http request timeout
url = "https://jsonplaceholder.typicode.com/posts"

try:
  response = requests.get(url, timeout=0.0001)
except requests.Timeout as error:
  print(error)
```
